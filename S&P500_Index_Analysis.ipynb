```python
import yfinance as yf
```

```python
#S&P 500 지수 데이터 다운로드
sp500 = yf.download('^GSPC', start='2000-01-01', end='2025-04-05')
print (sp500.head)
```

```python
import matplotlib.pyplot as plt

print (sp500.columns)

sp500["Return"] = sp500["Close"].pct_change()

#monthly = sp500["Return"].resample("M"),mean()

sp500["Close"].plot(figsize = (10,6), title = "S&P500 Monthly Avg Proit")
plt.grid(True)
plt.show()
# sp500['Adj Close'].plot(figsize=(10,6))
# plt.title('S&P 500 지수 가격 추이')
# plt.xlabel('Year')
# plt.ylabel('Price')
# plt.grid(True)
# plt.show()
```

```python
sp500['Daily Return']=sp500['Close'].pct_change()
```

```python
sp500['Daily Return'].hist(bins=50, figsize=(10,6))
plt.title('S&P500 Daily Return Graph')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
base_date = '2025-04-03'
base_price = sp500.loc[base_date]['Close']

sp500['Market Cap Estimate'] = (sp500['Close']/base_price)*45.388
```

```python
sp500['Market Cap Estimate'].plot(figsize=(10,6))
plt.title('S&P500 Index Cap Estimate')
plt.xlabel('Year')
plt.ylabel('Dollar(Trillion)')
plt.grid(True)
plt.show()
```
